```python
#dataframe with purchase value ans sales value and 5 different items 
s1=pd.Series(["pen","pencil","box","scale","peer"])
s2=pd.Series(np.random.randint(1,20,5))
s3=pd.Series(np.random.randint(1,20,5))
s4=pd.Series(np.random.randint(1,20,5)) 
s5=pd.Series(np.random.randint(1,20,5))
frame={"item":s1,"Cost price":s2,"Sale price":s3} 
df=pd.DataFrame(frame)
print(df)
```

```python
#find id profit or loss and add a new column 
df["P/F"]=df["Cost price"]-df["Sale price"]
print(df)
df["status"]=df["P/F"]>0 
print(df)
```

```python
# starting starts with p 
df[df["item"].str.startswith("p")]
```
